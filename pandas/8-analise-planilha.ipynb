```python
#Planeta com maior orbita_dias

maior_orbita = df[df['orbita_dias'] == df['orbita_dias'].max()]
maior_orbita
```

```python
#Relação entre Massa vs. Número de Satélites
#Correlação próxima a 1 significa que planetas mais massivos → mais satélites.

correlacao = df['massa_kg'].corr(df['satelites'])
correlacao
```
