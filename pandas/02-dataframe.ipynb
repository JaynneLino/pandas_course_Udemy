```python
dados_notas = {
    'Nome' : ['Alice', 'Bob', 'Charlie', 'Laize', 'Jessica', 'Paulo'],
    'Matemática' : [20, 13, 19, 18, 14, 19],
    'Física' :[17, 19, 13, 15, 17, 15],
    'Química' : [16, 17, 15, 14, 12, 16]
    }
```

```python
#Dataframe das notas

df_notas = pd.DataFrame(dados_notas)
df_notas
```

```python
#Calcular as médias das notas por aluno

#Criar uma nova coluna no DataFrame

df_notas['Média'] = df_notas[['Matemática', 'Física', 'Química']].mean(axis=1)
df_notas
```

```python
#Identificar alunos aprovados, com nota igual ou maior que 17.

alunos_aprovados = df_notas[df_notas['Média'] >= 17]
alunos_aprovados
```

```python
#Exibir apenas o nome e a média dos alunos aprovados com nota >= 17.

alunos_aprovados[['Nome', 'Média']]
```
